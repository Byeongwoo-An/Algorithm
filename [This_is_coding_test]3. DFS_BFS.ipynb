# 큐

```python
# 리스트보다 빠름
from collections import deque
queue = deque()

queue.append(5)
queue.append(2)
queue.append(3)
queue.append(7)
queue.popleft()
queue.append(1)
queue.append(4)
queue.popleft()

print(queue)
queue.reverse()
print(queue)
```

deque.append(item): item을 데크의 오른쪽 끝에 삽입한다.

deque.appendleft(item): item을 데크의 왼쪽 끝에 삽입한다.

deque.pop(): 데크의 오른쪽 끝 엘리먼트를 가져오는 동시에 데크에서 삭제한다.

deque.popleft(): 데크의 왼쪽 끝 엘리먼트를 가져오는 동시에 데크에서 삭제한다.

deque.extend(array): 주어진 배열(array)을 순환하면서 데크의 오른쪽에 추가한다.

deque.extendleft(array): 주어진 배열(array)을 순환하면서 데크의 왼쪽에 추가한다.

deque.remove(item): item을 데크에서 찾아 삭제한다.

deque.rotate(num): 데크를 num만큼 회전한다(양수면 오른쪽, 음수면 왼쪽).

# 재귀 함수

```python
def recursive_function():
    print("재귀 함수를 호출합니다.")
    recursive_function()
recursive_function()
```

```python
def recursive_function(i):
    if i == 100:
        return 
    print("재귀 함수를 호출합니다.")
    recursive_function(i+1)
recursive_function(1)
```

```python
def factorial_iterative(n):
    result = 1
    for i in range(1,n+1):
        result *= i
    return result

def factorial_recursive(n):
    if n<=1:
        return 1
    return n*factorial_recursive(n-1)

print(factorial_iterative(5))
print(factorial_recursive(5))
```

```python
graph_lst = [[] for _ in range(3)]

graph_lst[0].append((1, 7))
graph_lst[0].append((2, 5))

graph_lst[1].append((0,7))

graph_lst[2].append((0,5))

graph_lst
```

```python
def dfs(graph_dfs, v, visited):
    visited[v] = True
    print(v, end=' ')
    for i in graph_dfs[v]:
        if not visited[i]:
            dfs(graph_dfs, i, visited)

graph_dfs = [[],[2,3,8],[1,7],[1,4,5],[3,5],[3,4],[7],[2,6,8],[1,7]]
visited = [False]*9
dfs(graph_dfs,1,visited)
```

```python
from collections import deque

def bfs(graph_bfs, start, visited):
    
    que = deque([start])
    visited[start] = True
    while que:
        v = que.popleft()
        print(v, end=' ')
        for i in graph_bfs[v]:
            if not visited[i]:
                que.append(i)
                visited[i] = True
                
graph_bfs = [[],[2,3,8],[1,7],[1,4,5],[3,5],[3,4],[7],[2,6,8],[1,7]]
visited = [False]*9
bfs(graph_bfs,1,visited)
```

# 음료수 얼려 먹기

```python
# DFS
```

```python
n, m = map(int, input().split())
graph = []
for i in range(n):
    graph.append(list(map(int, input())))
```

```python
def dfs(x,y):
    if x<= -1 or x >= n or y<=-1 or y>=m:
        return False
    if graph[x][y] == 0:
        graph[x][y] = 1
        dfs(x-1,y)
        dfs(x,y-1)
        dfs(x+1,y)
        dfs(x,y+1)
        return True
    return False

result = 0
for i in range(n):
    for j in range(m):
        if dfs(i,j) == True:
            result += 1
print(result)
```

# 미로 탈출

```python
from collections import deque

n,m = map(int, input().split())
graph = []
for i in range(n):
    graph.append(list(map(int, input())))
```

```python
dx = [-1,1,0,0]
dy = [0,0,-1,1]

def bfs(x,y):
    que = deque()
    que.append((x,y))
    
    while que:
        x,y = que.popleft()
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            
            if nx < 0 or ny < 0 or nx >= n or ny >= m:
                continue
            if graph[nx][ny] == 0:
                continue
            if graph[nx][ny] == 1:
                graph[nx][ny] = graph[x][y] + 1
                que.append((nx,ny))
    return graph[n-1][m-1]
print(bfs(0,0))
```
